```python
from qiskit import QuantumCircuit
from qiskit.visualization.timeline import draw
import numpy as np
```

```python
def QFT(n):
    qc = QuantumCircuit(n)
    for i in range(n):
        # i target j is control
        qc.h(i)
        for j in (range(i+1,n)):
            qc.cp(np.pi/2**j,j,i)
        qc.barrier()
    qc.measure_all()
    return qc.draw('mpl')
```
